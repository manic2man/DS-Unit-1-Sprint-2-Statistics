```python
def tt_ind_df( left_dataframe, right_dataframe, shared_columns_list):
  for item in shared_columns_list:
    print(item)
    print('{:.10f}'.format(ttest_ind(left_dataframe[item], right_dataframe[item], nan_policy='omit').pvalue))
```
